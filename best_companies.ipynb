```python
# dependencies
import requests
import re
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# page to scrape
url = 'https://www.builtinchicago.org/awards/best-places-to-work-chicago'
```

```python
# collect page as text
html = requests.get(url)
```

```python
#create soup object
soup = bs(html.text,'html.parser')
```

```python
# narrow text search
companies = soup.select('.view-content')
```

```python
# create list of dictionaries that will make the dataframe

# holds company info dictionaries
best_companies = []

# website url
base_url ='https://www.builtinchicago.org'

# loop through each entry
for i in companies[1].select('.views-row'):
    
    # create dictionary of items
    item = {}
    item["title"] = i.select('.title > a')[0].text
    item["type"] = i.select('.field-type')[0].text
    item["description"] = i.select(".item")[0].text
    item["employees"] = i.select(".item")[1].text
    item["founded"] = i.select(".item")[2].text
    try:
        item["location"] = i.select(".item")[3].text
    except:
        item["location"] = "Unknown"
    try:
        item["openings"] = re.findall('[0-9]+', i.select(".open-jobs > a")[0].text)[0]
    except:
        item["openings"] = 0
    try:
        item["link"] = base_url + i.select(".open-jobs > a")[0]['href']
    except:
        item["link"] = "no link available"
    
    # add dictionary to list
    best_companies.append(item)
```

```python
# create dataframe
df = pd.DataFrame(best_companies)
df
```
